## Datos en Scikit-Learn

La gran mayora de las veces scikit-learn almacena los datos en arrays de 2 dimensiones; shape igual a [n_samples, n_features]. Algunos algortimos especficos aceptan matrices dispersas (**scipy.sparse**).

- **n_samples:** Nmero de ejemplos. Cada uno es un item a procesar y pueden ser un documento, una imagen, un sonido, un video, una fila de una base de datos o de un archivo CSV o cualquier cosa que pueda representarse con caractersticas cuantitativas.

- **n_features:** Nmero de caractersticas usadas para describir a cada item de forma cuantitativa. Por lo general son valores reales auqnue pueden ser categricas o valores discretos.

**NOTA:** Es posible que se pueda llegar a manejar millones de caractersticas, en donde puede llegar a ser cero en la mayoria de los casos, por lo cual es recomendable implementar matrices dispersas que tienen mejor manejo de memoria.

Los ejemplos se representaran en las filas mientras que las caractersticas corresponderan a las columnas (**dimensiones**)
