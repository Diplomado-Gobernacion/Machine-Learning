# Visualización de Datos en Python - Scikit-Learn

## Datos en Scikit-Learn

La gran mayoría de las veces scikit-learn almacena los datos en arrays de 2 dimensiones; shape igual a [n_samples, n_features]. Algunos algortimos específicos aceptan matrices dispersas (**scipy.sparse**).

- **n_samples:** Número de ejemplos. Cada uno es un item a procesar y pueden ser un documento, una imagen, un sonido, un video, una fila de una base de datos o de un archivo CSV o cualquier cosa que pueda representarse con características cuantitativas.

- **n_features:** Número de características usadas para describir a cada item de forma cuantitativa. Por lo general son valores reales auqnue pueden ser categóricas o valores discretos.

**NOTA:** Es posible que se pueda llegar a manejar millones de características, en donde puede llegar a ser cero en la mayoria de los casos, por lo cual es recomendable implementar matrices dispersas que tienen mejor manejo de memoria.

Los ejemplos se representaran en las filas mientras que las características corresponderan a las columnas (**dimensiones**)
